```python
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
data = pd.read_csv("dataset\diabetes.csv")
data = data.sample(frac = 1)
new_data = data.values
```

```python
new_data.shape
```

```python
total_data = new_data[:650,:8]
total_label = new_data[:650,8]
test_data = new_data[650:,:8]
test_label = new_data[650:,8]
```

```python
from keras import layers
from keras import models
```

```python
model = models.Sequential()
model.add(layers.Dense(256,input_shape = (8,)))
#model.add(layers.Dense(256,activation = "relu"))
model.add(layers.Dense(128,activation = "relu"))
model.add(layers.Dense(64,activation = "relu"))
model.add(layers.Dense(1,activation = "sigmoid"))
```

```python
model.compile(optimizer = "adam",loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
k = 5
sample = len(total_data)//k
validation_scores = []
train_accuracy = []
train_loss = []
for z in  range(k):
    validation_data = total_data[z * sample :(z + 1) * sample]
    validation_label = total_label[z * sample :(z + 1) * sample]
    train_data = np.concatenate([total_data[:z * sample] , total_data[(z + 1) * sample:]])
    train_label = np.concatenate([total_label[:z * sample] , total_label[(z + 1) * sample:]])
    history = model.fit(train_data,train_label, epochs = 200)
    validation_score = model.evaluate(validation_data,validation_label)
    validation_scores.append(validation_score)
    train_accuracy.append(history.history["accuracy"])
    train_loss.append(history.history["loss"])
```

```python
new_validation_data = total_data[520:]
new_train_data = total_data[:520]
new_validation_label = total_label[520:]
new_train_label = total_label[:520]
```

```python
new_train_label = new_train_label.astype("int32")
new_validation_label = new_validation_label.astype("int32")
```

```python
history2 = model.fit(new_train_data, new_train_label, epochs = 200, validation_data = (new_validation_data,new_validation_label))
```

```python
t_accuracy = history2.history["accuracy"]
t_loss = history2.history["loss"]
v_accuracy = history2.history["val_accuracy"]
v_loss = history2.history["val_loss"]
epochs = np.arange(1,len(t_accuracy) + 1,1)
```

```python
plt.title("Train and Validation Accuracy")
plt.plot(epochs, t_accuracy, "bo", label = "Train Accuracy")
plt.plot(epochs, v_accuracy, "b", label = "Validation Accuracy")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
plt.title("Train and Validation Loss")
plt.plot(epochs, t_loss,"bo",label = "Train Loss")
plt.plot(epochs, v_loss,"b",label = "Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
