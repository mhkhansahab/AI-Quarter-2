```python
model = models.Sequential()
model.add(layers.Dense(units = 4,input_shape = [1]))
model.add(layers.Dense(units = 1))
```

```python
model.compile(loss = "mean_squared_error", optimizer = keras.optimizers.Adam(0.1), metrics=["accuracy"])
```
