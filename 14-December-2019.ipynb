```python
import numpy as np
import matplotlib.pylab as plt
```

```python
X = np.array([2,4,6,8,10]) #input
Y = np.array([5,10,15,20,25]) #output
m = Y/X
slope = m.mean()
print("Slope",slope)
x = 6
plt.linspace(x,y,100)
y = slope * x
print("Value of y: ",y)
```

```python
plt.plot(X,Y)
plt.plot(x,y)
```

```python
plt.scatter(X,Y)
```

```python
X = np.array([2,4,6,8,10]) #input
Y = np.array([5,10,20,22,27]) #output
m = Y/X
slope = m.mean()
print("Slope",slope)
start = X[0]
end = X[-1]
x = plt.linspace(start,end,100)
y = slope * x
```

```python
plt.scatter(X,Y)
plt.plot(x,y)
```
